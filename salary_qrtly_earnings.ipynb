```python
#changing column names to reflect quarters so they can be seen horizontally by sector
salaryq1_df.rename(columns= { 'qtrly_estabs':'q1_establishments', 'total_qtrly_wages':'q1_total_wages', 'avg_wkly_wage':'q1_avg_wages', 'lq_qtrly_estabs':'q1_lq_establishments', 'lq_total_qtrly_wages':'q1_lq_total_wages', 'oty_qtrly_estabs_chg':'q1_establishments_chg'}, inplace=True)
salaryq2_df.rename(columns= { 'qtrly_estabs':'q2_establishments', 'total_qtrly_wages':'q2_total_wages', 'avg_wkly_wage':'q2_avg_wages', 'lq_qtrly_estabs':'q2_lq_establishments', 'lq_total_qtrly_wages':'q2_lq_total_wages', 'oty_qtrly_estabs_chg':'q2_establishments_chg'}, inplace=True)
salaryq3_df.rename(columns= { 'qtrly_estabs':'q3_establishments', 'total_qtrly_wages':'q3_total_wages', 'avg_wkly_wage':'q3_avg_wages', 'lq_qtrly_estabs':'q3_lq_establishments', 'lq_total_qtrly_wages':'q3_lq_total_wages', 'oty_qtrly_estabs_chg':'q3_establishments_chg'}, inplace=True)
salaryq4_df.rename(columns= { 'qtrly_estabs':'q4_establishments', 'total_qtrly_wages':'q4_total_wages', 'avg_wkly_wage':'q4_avg_wages', 'lq_qtrly_estabs':'q4_lq_establishments', 'lq_total_qtrly_wages':'q4_lq_total_wages', 'oty_qtrly_estabs_chg':'q4_establishments_chg'}, inplace=True)
```

```python
#keeping only codes for sectors considered to be arts / design related
desired_codes = ['711110', '711120', '711130', '711190', '711410', '711510', '712110', '712120', '512110' ,'512191' ,'512230' ,'512240', '512250', '512290']
salaryq1_df = salaryq1_df[salaryq1_df['industry_code'].isin(desired_codes)]     
salaryq2_df = salaryq2_df[salaryq2_df['industry_code'].isin(desired_codes)]
salaryq3_df = salaryq3_df[salaryq3_df['industry_code'].isin(desired_codes)]
salaryq4_df = salaryq4_df[salaryq4_df['industry_code'].isin(desired_codes)]

```

```python
#resolving issue with doubled museum rows
salaryq1_df.loc[177, 'industry_code'] = '7121100'
```

```python
#resolving issue with doubled museum rows
salaryq4_df.loc[177, 'industry_code'] = '7121100'
```

```python
#changing industry code from string to int 
salaryq1_df['industry_code']= salaryq1_df['industry_code'].astype(int)
salaryq2_df['industry_code']= salaryq2_df['industry_code'].astype(int)
salaryq3_df['industry_code'] = salaryq3_df['industry_code'].astype(int)
salaryq4_df['industry_code'] = salaryq4_df['industry_code'].astype(int)
```

```python
#merging process
salary_total_df = pd.merge(salaryq1_df, salaryq2_df, on='industry_code', how='outer')
```

```python
salary_total_df = pd.merge(salary_total_df, salaryq3_df, on='industry_code', how='outer')
```

```python
salary_total_df = pd.merge(salary_total_df, salaryq4_df, on='industry_code', how='outer')
```

```python
#adding labels to industry codes
code_to_sector = { 711110: 'Theater Companies and Dinner Theaters', 711120: 'Dance Companies', 711130: 'Musical Groups and Artists', 711190: 'Other Performing Arts Companies', 711410: 'Agents and Managers of Performing Arts, Sports, and Similar Events', 711510: 'Independent Artists, Writers, and Performers', 712110: 'Museums', 7121100: 'Museums', 712120: 'Historical Sites', 512110: 'Motion Picture and Video Production', 512191: 'Teleproduction and Other Postproduction Services', 512230: 'Music Publishers', 512240: 'Sound Recording Studios', 512250: 'Record Production', 512290: 'Other Sound Recording Industries'}

salary_total_df['sector'] = salary_total_df['industry_code'].map(code_to_sector)
```

```python
#changing order so sector is next to industry code
new_order = ['industry_code','sector', 'q1_establishments', 'q1_total_wages', 'q1_avg_wages',
       'q1_lq_establishments', 'q1_lq_total_wages',
       'q2_establishments', 'q2_total_wages', 'q2_avg_wages',
       'q2_lq_establishments', 'q2_lq_total_wages',
       'q3_establishments', 'q3_total_wages', 'q3_avg_wages',
       'q3_lq_establishments', 'q3_lq_total_wages',
       'q4_establishments', 'q4_total_wages', 'q4_avg_wages',
       'q4_lq_establishments', 'q4_lq_total_wages']
```

```python
salary_total_df = salary_total_df[new_order]

#resetting index
salary_total_df = salary_total_df.reset_index(drop=True)
```

```python
#moving the first museum row down to be with other museum row 
row_to_move = salary_total_df.loc[[0]]
```

```python
salary_total_df = salary_total_df.drop(0)
```

```python
#defining location for moving row 
row_1 = salary_total_df.loc[:13]
row_2 = salary_total_df.loc[14:]
```

```python
salary_total_df = pd.concat([row_1, row_to_move, row_2])

#resetting index
salary_total_df = salary_total_df.reset_index(drop=True)
```

```python
#dropping historical sites because there's not enough data there for wages
salary_total_df = salary_total_df.drop(14)
```

```python
salary_total_df['q1_avg_monthly_wages'] = salary_total_df['q1_avg_wages'] * 4 
salary_total_df['q2_avg_monthly_wages'] = salary_total_df['q2_avg_wages'] * 4 
salary_total_df['q3_avg_monthly_wages'] = salary_total_df['q3_avg_wages'] * 4 
salary_total_df['q4_avg_monthly_wages'] = salary_total_df['q4_avg_wages'] * 4 
salary_total_df['total_mdn_monthly_wages'] = salary_total_df[['q1_avg_monthly_wages', 'q2_avg_monthly_wages','q3_avg_monthly_wages', 'q4_avg_monthly_wages']].median(axis=1) 
salary_total_df['mdn_lq_total_wages'] = salary_total_df[['q1_lq_total_wages', 'q2_lq_total_wages', 'q3_lq_total_wages', 'q4_lq_total_wages']].median(axis=1)
salary_total_df['mdn_lq_establishments'] = salary_total_df[['q1_lq_establishments','q2_lq_establishments', 'q3_lq_establishments', 'q4_lq_establishments']].median(axis=1)
```

```python
salary_total_df = salary_total_df.drop(columns=['q1_lq_establishments','q2_lq_establishments', 'q3_lq_establishments', 'q4_lq_establishments','q1_lq_total_wages', 'q2_lq_total_wages', 'q3_lq_total_wages', 'q4_lq_total_wages', 'q1_total_wages','q2_total_wages','q3_total_wages', 'q4_total_wages'])
```

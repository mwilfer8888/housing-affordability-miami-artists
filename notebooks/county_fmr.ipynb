```python
import pandas as pd 
```

```python
#preparing excel worksheet for import 
worksheet = 'fy23-fmr-zipcode.xlsx'
sheet_name = 'SAFMRs'
```

```python
#reading overall df from excel sheet, this data can be used for counties all around the US if needed
fmr_df = pd.read_excel(worksheet, sheet_name=sheet_name)
fmr_df
```

```python
#creating new dataframe just for zip codes within Miami-Dade County 
miami_fmr_df = fmr_df[fmr_df['HUD Metro Fair Market Rent Area Name'] == 'Miami-Miami Beach-Kendall, FL HUD Metro FMR Area']
```

```python
#dropping the 90% and 110% payment standard columns as they are not needed for this project
miami_fmr_df = miami_fmr_df.drop(columns=['HUD Metro Fair Market Rent Area Name', 'HUD Area Code','SAFMR\n0BR - 90%\nPayment\nStandard', 'SAFMR\n0BR - 110%\nPayment\nStandard','SAFMR\n1BR - 90%\nPayment\nStandard',
       'SAFMR\n1BR - 110%\nPayment\nStandard','SAFMR\n2BR - 90%\nPayment\nStandard',
       'SAFMR\n2BR - 110%\nPayment\nStandard','SAFMR\n3BR - 90%\nPayment\nStandard',
       'SAFMR\n3BR - 110%\nPayment\nStandard','SAFMR\n4BR - 90%\nPayment\nStandard',
       'SAFMR\n4BR - 110%\nPayment\nStandard'], axis=1)
```

```python
#renaming columns for easier understanding
miami_fmr_df = miami_fmr_df.rename(columns= {'ZIP\nCode': 'zip_code', 
                                             'SAFMR\n0BR': 'studio', 
                                             'SAFMR\n1BR': '1BR',
                                            'SAFMR\n2BR': '2BR', 
                                            'SAFMR\n3BR': '3BR', 
                                            'SAFMR\n4BR': '4BR'})
```

```python
#saving a csv file with just rental prices
miami_fmr_df.to_csv('miami_fmr.csv', index=False)
```

```python
#now adding calculations related to housing affordability 

#this calculation is based on the 30% rule, which states that a household should not spend more than 30% of their income on housing
miami_fmr_df['studio_monthly_income_rq'] = miami_fmr_df['studio'] / 0.30
miami_fmr_df['1BR_monthly_income_rq'] = miami_fmr_df['1BR'] / 0.30
miami_fmr_df['2BR_monthly_income_rq'] = miami_fmr_df['2BR'] / 0.30
miami_fmr_df['3BR_monthly_income_rq'] = miami_fmr_df['3BR'] / 0.30
miami_fmr_df['4BR_monthly_income_rq'] = miami_fmr_df['4BR'] / 0.30

# this calculates the yearly income required for the different types of housing 
miami_fmr_df['studio_yrly_income_rq'] = miami_fmr_df['studio_monthly_income_rq'] * 12
miami_fmr_df['1BR_yrly_income_rq'] = miami_fmr_df['1BR_monthly_income_rq'] * 12
miami_fmr_df['2BR_yrly_income_rq'] = miami_fmr_df['2BR_monthly_income_rq'] * 12
miami_fmr_df['3BR_yrly_income_rq'] = miami_fmr_df['3BR_monthly_income_rq'] * 12
miami_fmr_df['4BR_yrly_income_rq'] = miami_fmr_df['4BR_monthly_income_rq'] * 12

# this calculates the hourly wage required to afford the different types of housing
miami_fmr_df['studio_hourly_wage_rq'] = miami_fmr_df['studio_yrly_income_rq'] / 2080
miami_fmr_df['1BR_hourly_wage_rq'] = miami_fmr_df['1BR_yrly_income_rq'] / 2080
miami_fmr_df['2BR_hourly_wage_rq'] = miami_fmr_df['2BR_yrly_income_rq'] / 2080
miami_fmr_df['3BR_hourly_wage_rq'] = miami_fmr_df['3BR_yrly_income_rq'] / 2080
miami_fmr_df['4BR_hourly_wage_rq'] = miami_fmr_df['4BR_yrly_income_rq'] / 2080

```

```python
#converting all calculations to integers
calculations = ['studio_monthly_income_rq', '1BR_monthly_income_rq',
       '2BR_monthly_income_rq', '3BR_monthly_income_rq',
       '4BR_monthly_income_rq', 'studio_yrly_income_rq', '1BR_yrly_income_rq',
       '2BR_yrly_income_rq', '3BR_yrly_income_rq', '4BR_yrly_income_rq',
       'studio_hourly_wage_rq', '1BR_hourly_wage_rq', '2BR_hourly_wage_rq',
       '3BR_hourly_wage_rq', '4BR_hourly_wage_rq']

for col in calculations:
    miami_fmr_df[col] = miami_fmr_df[col].astype(int)
```

```python
#resetting index
miami_fmr_df = miami_fmr_df.reset_index(drop=True)
miami_fmr_df
```

```python
#reordering columns for better understanding
miami_fmr_df = miami_fmr_df[['zip_code', 'studio', 'studio_monthly_income_rq', 'studio_yrly_income_rq', 'studio_hourly_wage_rq',
                             '1BR', '1BR_monthly_income_rq', '1BR_yrly_income_rq', '1BR_hourly_wage_rq',
                             '2BR', '2BR_monthly_income_rq', '2BR_yrly_income_rq', '2BR_hourly_wage_rq',
                             '3BR', '3BR_monthly_income_rq', '3BR_yrly_income_rq', '3BR_hourly_wage_rq',
                             '4BR', '4BR_monthly_income_rq', '4BR_yrly_income_rq', '4BR_hourly_wage_rq']]
```

```python
#saving as its own dataframe 
miami_fmr_df.to_csv('miami_fmr_df_with_calculations.csv', index=False)
```

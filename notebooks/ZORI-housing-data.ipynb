```python
import pandas as pd 
```

```python
zori = 'Zip_zori_uc_sfrcondomfr_sm_month.csv'
```

```python
zori_rental_df = pd.read_csv(zori)
zori_rental_df = pd.DataFrame(zori_rental_df)
```

```python
zori_rental_df = zori_rental_df[zori_rental_df['State'] == 'FL']
zori_rental_df
```

```python
miami_rental_df = zori_rental_df[zori_rental_df['CountyName'] == 'Miami-Dade County']
miami_rental_df
```

```python
miami_rental_df.columns
```

```python
miami_rental_df = miami_rental_df.drop(columns=['RegionID', 'SizeRank', 'RegionType', 'StateName', 'State', 'Metro', 'CountyName', 'City'])
```

```python
miami_rental_df.rename(columns={'RegionName':'zip_code'}, inplace=True)
```

```python
miami_rental_df = miami_rental_df.fillna(0)
```

```python
miami_rental_df.reset_index(drop=True, inplace=True)
miami_rental_df
```

```python
exclude_col = 'zip_code'

miami_rental_df.columns = [pd.to_datetime(col, errors='coerce').date() if col != exclude_col else col for col in miami_rental_df.columns]
```

```python
miami_rental_df
```

```python
miami_rental_df = miami_rental_df.T
```

```python
miami_rental_df
```

```python
miami_rental_df.columns = miami_rental_df.iloc[0]
```

```python
miami_rental_df
```

```python
miami_rental_df = miami_rental_df.astype(int)
```

```python
miami_rental_df
```

```python
new_header = miami_rental_df.iloc[0] #grab the first row for the header
miami_rental_df = miami_rental_df[1:] #take the data less the header row
miami_rental_df.columns = new_header #set the header row as the df header
```

```python
miami_rental_df
```

```python
#maybe this dataset can be used as a time-series 
miami_rental_df.to_csv('miami-rentals-over-time-zipcode.csv', index=True)
```
